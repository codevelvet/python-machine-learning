In this notebook, we will read a csv file

We will create a dataframe and clean up the data and save the dataframe

```python
import pandas as pd
```

```python
df = pd.read_csv("Titanic.csv")
```

```python
df.head()
```

### Creating a brand new column called family_size

family_size = Parch + SibSp + 1

```python
# creating a brand new column called family_size = Parch + SibSp + 1

df["family_size"] = df["Parch"] + df["SibSp"] + 1
```

```python
df.head()
```

### Convert columns Gender and Embarked into numeric 

```python
gen = {"male": 0, "female": 1}

df["Gender"] = df["Gender"].map(gen)
```

```python
df.head()
```

### Addressing missing values in Embarked column

```python
df.isnull().sum()
```

```python
df["Embarked"].value_counts()
```

Notice that port "S" is the most frequent value for Embarked.

We will use "S" to fill null values in this column.

                                    

```python
df["Embarked"].fillna("S", inplace=True)
```

```python
df.isnull().sum()
```

### Transforming Embarked categorical feature into numeric

We are transforming using a dictionary and map() method.

```python
emb = {"S": 0, "C":1, "Q":2}

df["Embarked"] = df["Embarked"].map(emb)
```

```python
df.head()
```

### Now let us save this dataframe

Saving it as titanic.pk

```python
df.to_pickle("titanic.pk")
```
